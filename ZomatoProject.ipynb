```python
country_name=final_df["Country"].value_counts().index[:3]
country_transaction=final_df["Country"].value_counts().values[:3]
centre_circle = plt.Circle((0,0),0.70,fc='white')
plt.figure(figsize=(8,6))
plt.pie(country_transaction,labels=country_name,
        autopct="%1.2f%%",explode = (0.05,0.05,0.05),
        pctdistance=1.2,labeldistance=1.6,
        colors = ['#ff9999','#66b3ff','#99ff99'])
fig = plt.gcf()
fig.gca().add_artist(centre_circle)
plt.title("TOP 3 COUNTRIES")
plt.show()
```

```python
rating=final_df.groupby(["Aggregate rating","Rating color","Rating text"]).size().reset_index().rename(columns={0:"Rating Count"})
plt.figure(figsize=(12,4))
sns.barplot(x="Aggregate rating",y="Rating Count",data=rating,palette="viridis")
plt.xticks(rotation=90)
plt.title("RATINGS")
plt.xlabel("RATING")
plt.ylabel("COUNT")
plt.show()
```

```python
plt.figure(figsize=(10,4))
sns.countplot(x="Rating text",data=final_df, palette='cubehelix')
plt.show()
```

```python
countries=final_df["Cuisines"].value_counts().index[:10]
ordersc=final_df["Cuisines"].value_counts().values[:10]
plt.figure(figsize=(12,6))
plt.xticks(rotation=90)
sns.barplot(y=ordersc,x=countries,palette="flare")
plt.title("TOP 10 CUISINES")
plt.show()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Extract top 10 countries based on aggregate ratings
Country = final_df["Country"].value_counts().index[:10]
rating = final_df["Aggregate rating"].value_counts().values[:10]

# Plotting the bar chart
plt.title('Top 10 countries based on Aggregate Ratings')
sns.barplot(x=Country, y=rating, palette="flare")
plt.xlabel('Country')
plt.ylabel('Number of Ratings')
plt.xticks(rotation=45)  # Rotate x-axis labels for better readability
plt.show()
```
